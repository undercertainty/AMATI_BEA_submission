# AMaTi Stopword Generator

Very quick script to generate an encoding of German stopwords in Amati (ie. clingo) format. Details of the originals are in the files.

I'm also adding the separator `--` which seems to be particularly common in the training files.

```python
with open('german_stopwords_plain.txt') as fIn:
    with open('german_stopwords.amati', 'w') as fOut:
        for nl in fIn.readlines():
            if nl[0]!=';':
                fOut.write(f'stopword("{nl.strip()}").\n')
        fOut.write(f'stopword("--").\n')
        
```
